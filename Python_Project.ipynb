##### *Q.8 Write the code to find the percentage missing value (column-wise) in master_dataframe. Drop unnecessary columns.

```python
# getting all the null values using isnull function and adding them using sum function.
Null_values = master_df.isnull().sum().sort_values(ascending=False)

# Extracting percentage of null values in master_df, summing all the null values divided by total number of observation in master_df. 
Percentage = round((master_df.isnull().sum()/len(master_df))*100, 2).sort_values(ascending=False)

# Concatinating both total and percentage of null in single dataframe for easy viewing
missing_data = pd.concat([Null_values, Percentage], axis=1, keys=['Null_values Total', 'Percentage'])
```

##### *Q. 13 Using the user defined function convert the ‘rasied_amount_used’ column into a categorical column as follow*
##### *a) If the amount is less than 5 Millions then recode as TypeA.*
##### *b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.*
##### *c) If the amount is greater than 7 Millions then recode as TypeC.*

```python
# using user define function recoding data as per above.

def type(x):
    if x < 5000000:
        return 'TypeA'
    elif x >= 5000000 and x <=7000000 :
        return 'TypeB'
    else:
        return 'TypeC'
```

```python
master_df['raised_amount_usd'] = master_df['raised_amount_usd'].apply(type)
master_df.head()
```

```python
# converting raised amount coulmn into categorical

master_df['raised_amount_usd'] = master_df['raised_amount_usd'].astype('category')
```
